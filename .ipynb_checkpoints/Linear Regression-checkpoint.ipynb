```python
import numpy as np
import matplotlib.pyplot as plt

# pizza sizes in inches
pizza_size = [6, 8, 12, 14, 18]
#pizza prices according to the above the sizes
pizza_price = [350, 775, 1150, 1395, 1675]

fit = np.polyfit(pizza_size, pizza_price,1)
fit_fn = np.poly1d(fit)
# fit_fn >>> poly1d([ 107.21491228, -174.69298246])
# y = 107.21x - 174.69, this is our regression line/best fit line
plt.figure(figsize=(10, 5), dpi=90)
plt.plot(pizza_size, pizza_price, "yo", pizza_size, fit_fn(pizza_size), "--k")
plt.xlim(0, 20)
plt.ylim(0, 1800)
```

```python
import matplotlib.pyplot as plt


def best_fit(X, Y):
    n = len(X)
    xbar = sum(X)/n
    ybar = sum(Y)/n
    xybar = sum([xi*yi for xi, yi in zip(X, Y)])/n
    xbar_square = xbar**2
    x_square_bar = sum([xi**2 for xi in X])/n
    
    m = ((xbar * ybar) - xybar)/(xbar_square - x_square_bar)
    c = ybar - m*xbar
    
    print(f"Equation of best fit line: y = {m:.2f}x + ({c:.2f})")
    
    return m, c, ybar

# pizza sizes in inches
pizza_size = [6, 8, 12, 14, 18]
#pizza prices according to the above the sizes
pizza_price = [350, 775, 1150, 1395, 1675]

m, c, ybar = best_fit(pizza_size, pizza_price)
yfit = [m*xi + c for xi in pizza_size]
r_squared = sum([(i-ybar)**2 for i in yfit])/sum([(i-ybar)**2 for i in pizza_price])
print(f"R-Squared: {r_squared:.3f}")

plt.figure(figsize=(12,6), dpi=96)
plt.title("Linear regression")
plt.xlabel("Pizza Size (inches)")
plt.ylabel("Pizza Price")
plt.scatter(pizza_size, pizza_price)
plt.plot(pizza_size, yfit)
```
